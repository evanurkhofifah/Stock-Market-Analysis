#### Moving Average (MA) dan Moving Average Convergence Divergence (MACD)

Moving Average indikator untuk menghitung pergerakan harga rata-rata dari suatu saham dalam suatu rentang waktu.
- Jika harga saham berada di atas nilai MA, maka harga saham tersebut sedang dalam tren naik
- Jika harga berada di bawah MA, maka harga saham sedang dalam tren turun.
- Jika MA 50 (jangka pendek) di atas MA 200 (jangka panjang), maka harga saham tersebut sedang dalam tren naik,
- Jika MA 50 berada di bawah MA 200, maka harga saham sedang dalam tren turun.

```python
#moving average 50 dan 200
df2['MA_50'] = df2['Close'].rolling(window=50).mean()
df2['MA_200'] = df2['Close'].rolling(window=200).mean()
```

##### MACD

- Jika garis MACD memotong ke atas garis sinyal (MACD Golden Cross) = Sinyal beli saham
- Jika garis MACD memotong ke bawah garis sinyal (MACD Dead Cross) = Sinyal jual saham

```python
#Indikator Moving Average Convergence Divergence
EMA_12 = pd.Series(df2['Close'].ewm(span=12, min_periods=12).mean())
EMA_26 = pd.Series(df2['Close'].ewm(span=26, min_periods=26).mean())
MACD = pd.Series(EMA_12 - EMA_26)
MACD_signal = pd.Series(MACD.ewm(span=9, min_periods=9).mean())

fig = make_subplots(rows=2, cols=1)
fig.add_trace(go.Scatter(x=df2.index, y=df2.Close, name='Close'), row=1, col=1)
fig.add_trace(go.Scatter(x=df2.index, y=EMA_12, name='EMA 12'), row=1, col=1)
fig.add_trace(go.Scatter(x=df2.index, y=EMA_26, name='EMA 26'), row=1, col=1)
fig.add_trace(go.Scatter(x=df2.index, y=MACD, name='MACD'), row=2, col=1)
fig.add_trace(go.Scatter(x=df2.index, y=MACD_signal, name='Signal line'), row=2, col=1)

fig.update_layout(
    title_text="Moving Average Convergence Divergence",
    plot_bgcolor='rgba(37,37,37,1)',
    hovermode='x'
)

fig.show()
```

#### Bollinger Bands

- Jika harga menyentuh Upper Band dan harga penutupan masih di bawah Upper Band, maka kondisi dinyatakan overbought.
- Jika kondisi harga sudah menyentuh Lower Band, tetapi masih ditutup di atas Lower Band, dinyatakan sebagai oversold (sinyal untuk menjual atau membeli.
- Jika volatilitas sedang tinggi, maka jarak kedua band akan makin melebar, yang biasanya terjadi ketika perubahan kondisi sideways ke trending.
- Jika volatilitas pasar rendah maka jarak kedua band makin menyempit, dan biasanya terjadi ketika ada perubahan dari kondisi pasar trending menjadi sideways.

```python
# bollinger bands
df2['STD_50'] = df2['Close'].rolling(window=50).std()
#menghitung upper dan lower band
upper_bb = df2['MA_50'] + (df2['STD_50'] * 2)
lower_bb = df2['MA_50'] - (df2['STD_50'] * 2)
```

```python
#visualisasi bollinger bands
fig = go.Figure()

upBB  = go.Scatter(
        x=df2.index,
        y=upper_bb,
        name = 'Upper BB',
        hovertemplate = None
    )

lowBB  = go.Scatter(
        x=df2.index,
        y=lower_bb,
        name = 'Lower BB',
        hovertemplate = None
    )

data_bb = [ma_close, upBB, ma50, lowBB]

layout = go.Layout({
    'title': {
        'text': 'Bollinger Bands',
        'font': {
            'size': 25
        },
    },
},
hovermode="x unified", plot_bgcolor='rgba(37,37,37,1)'
#paper_bgcolor='rgba(37,37,37,1)
)

fig = go.Figure(data=data_bb, layout=layout)
fig.show()
```

```python
#visualisasi bollinger bands dengan candlestcik
bb_up = {
    'x': df2.index,
    'y': upper_bb,
    'type': 'scatter',
    'mode': 'lines',
    'line': {
        'width': 1,
        'color': 'dodgerblue'
            },
    'name': 'UPPER BB'
}

bb_low = {
    'x': df2.index,
    'y': lower_bb,
    'type': 'scatter',
    'mode': 'lines',
    'line': {
        'width': 1,
        'color': 'yellow'
            },
    'name': 'LOWER BB'
}

dataBB = [set_df, bb_up, cdl_ma50 , bb_low]
layout = go.Layout({
    'title': {
        'text': 'Bollinger Bands',
        'font': {
            'size': 25
        },
    },
}, plot_bgcolor='rgba(37,37,37,1)')


fig = go.Figure(data=dataBB, layout=layout)
fig.show()
```

#### Relative Strength Index (RSI)

RSI untuk mengetahui apakah suatu harga sudah overbought atau oversold.
- Jika RSI di atas 70 artinya pasar sudah overbought (jenuh beli) sehingga ada potensi turun, saatnya untuk jual.
- Jika RSI di bawah 30 artinya pasar sudah oversold (jenuh jual) sehingga ada potensi naik, saatnya untuk beli.

```python
# Returns RSI values
def rsi(close, periods = 14):

    close_delta = close.diff()

    # Make two series: one for lower closes and one for higher closes
    up = close_delta.clip(lower=0)
    down = -1 * close_delta.clip(upper=0)

    ma_up = up.ewm(com = periods - 1, adjust=True, min_periods = periods).mean()
    ma_down = down.ewm(com = periods - 1, adjust=True, min_periods = periods).mean()

    rsi = ma_up / ma_down
    rsi = 100 - (100/(1 + rsi))
    return rsi
```

```python
# Call RSI function from the talib library to calculate RSI
df2['RSI'] = rsi(df2['Close'])
```

#### Stochastic

```python
# Define periods
k_period = 14
d_period = 3

# Adds a "n_high" column with max value of previous 14 periods
n_high = df2['High'].rolling(k_period).max()

# Adds an "n_low" column with min value of previous 14 periods
n_low = df2['Low'].rolling(k_period).min()
```

```python
# Uses the min/max values to calculate the %k (as a percentage)
df2['%K'] = (df2['Close'] - n_low) * 100 / (n_high - n_low)

# Uses the %k to calculates a SMA over the past 3 values of %k
df2['%D'] = df2['%K'].rolling(d_period).mean()
```
